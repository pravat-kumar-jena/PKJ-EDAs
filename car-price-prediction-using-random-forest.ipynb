```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../input/car-price-prediction/CarPrice_Assignment.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df[['enginetype']].info()
```

```python
df['enginetype']=df['enginetype'].replace({'dohc':0, 'ohcv':1,'ohc':2, 'l':3, 'rotor':4, 'ohcf':5, 'dohcv':6})
```

```python
df['fueltype'].unique()
```

```python
df['fueltype']=df['fueltype'].replace({'gas':0, 'diesel':1})
```

```python
df['doornumber'].unique()
```

```python
df['doornumber']=df['doornumber'].replace({'two':2, 'four':4})
```

```python
df['carbody'].unique()
```

```python
df['carbody']=df['carbody'].replace({'convertible':1, 'hatchback':2, 'sedan':3, 'wagon':4, 'hardtop':5})
```

```python
df['drivewheel'].unique()
```

```python
df['drivewheel']=df['drivewheel'].replace({'rwd':0, 'fwd':1, '4wd':2})
```

```python
x = df[['fueltype','doornumber','carbody','drivewheel','enginetype','horsepower','peakrpm']]
```

```python
y = df['price']
```

```python
from sklearn.preprocessing import scale
```

```python
cols = x.columns
x = pd.DataFrame(scale(x))
x.columns = cols
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred)
print("r2_score", r2_score*100)
```
