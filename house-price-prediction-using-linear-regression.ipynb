```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib as plt
```

```python
df=pd.read_csv
```

```python
df[df['total_sqft'].str.contains('[Acres, Sq Meter, Yad, Cn, Guh, o, P]')==True]
```

```python
df['total_sqft']=df['total_sqft'].apply(lambda x : (float(x.split('-')[0])+float(x.split('-')[0])/2) if x.find('-')>0 else x)
df['total_sqft']=df['total_sqft'].astype(str).apply(lambda x : float(x.replace('Sq. Meter', ''))*10.76 if x.find('M')>0 else x)
df['total_sqft']=df['total_sqft'].astype(str).apply(lambda x : float(x.replace('Acres', ''))*43560.00 if x.find('A')>0 else x)
df['total_sqft']=df['total_sqft'].astype(str).apply(lambda x : float(x.replace('Sq. Yards', ''))*9.00 if x.find('Y')>0 else x)
df=df[df['total_sqft'].str.contains('[Yad, Cn, Guh, o, P]')==False]
df['total_sqft']=df['total_sqft'].astype(float)
```

```python
x = df[['total_sqft']]
y = df['price']
x
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
x = float(input("Area of the land in sqft"))
Land_Area_IN_sqft = [[x]]
predicted_result = model.predict(Land_Area_IN_sqft)
print(predicted_result)
```
