Popularity based anime recomendation system

```python
rating_with_name = rating.merge(anime,on = 'anime_id')
```

```python
rating_with_name
```

```python
 rating_with_name = rating_with_name.drop(['rating_x'], axis = 1)
```

```python
rating_with_name
```

```python
rating_with_name.isnull().sum()
```

```python
rating_with_name.shape
```

```python
# number of ratings
num_rating_df = rating_with_name.groupby('name').count()['rating_y'].reset_index()
num_rating_df.rename(columns={'rating_y':'num_ratings'},inplace=True)
```

```python
num_rating_df
```

```python
# Average rating
avg_rating_df= rating_with_name.groupby('name').mean()['rating_y'].reset_index()
avg_rating_df.rename(columns={'rating_y':'avg_ratings'},inplace=True)
```

```python
avg_rating_df 
```

```python
# Mearge the table
popular_df = num_rating_df.merge(avg_rating_df,on = 'name')
```

```python
popular_df
```

```python
popular_df[popular_df['num_ratings']>=250]
```

```python
# short on average rating in descnding order
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('num_ratings',ascending=False).head(50)
```

```python
popular_df.merge(anime,on='name')
```

```python
popular_df = popular_df.merge(anime,on='name')[['name','genre','type','episodes','num_ratings','avg_ratings']]
```
