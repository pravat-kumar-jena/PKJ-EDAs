```python
df['Job Roles']=df['Job Roles'].replace({'Android':1, 'Backend':2, 'Database':3, 'Frontend':4, 'IOS':5, 'Java':6, 'Mobile':7, 'SDE':8, 'Python':9, 'Web':10, 'Testing':11})
```

```python
df[['Job Roles']].info()
```

```python
#employment_status = set(df['Employment Status'])
df['Employment Status'] = df['Employment Status'].replace({'Full Time' : 0, 'Intern' : 1, 'Contractor' : 2, 'Trainee' : 3})
```

```python
df[['Employment Status']].info()
```

```python
x = df[['Rating','Job Roles','Employment Status']]
y = df['Salary']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split

```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state=0)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred)
print("r2score:",r2_score*100)
```

```python
rating = float(input("enter a rating"))
job_roles = int(input("enter a job roll no."))
employment_status = int(input("enter employment status"))
salary = [[rating, job_roles, employment_status]]
result = model.predict(sc.transform(salary))
print("salary =",result)
```
